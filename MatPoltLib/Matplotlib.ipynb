```python
x=[1,2,3,4,5,6,7]
y=[1000,2200,3400,2300,1200,4500,6000]
plt.plot(x,y,color='blue',linestyle='--',linewidth=2,marker='o',label='2025 Sales Data')
plt.title("Sales of the week")
plt.xlabel("Months")
plt.ylabel("Sales per day")
plt.legend(loc='upper left', fontsize=12)
plt.grid(color='gray',linestyle=':',linewidth=1)
plt.xlim(1,7)
plt.ylim(0,7000)
plt.xticks([1,2,3,4,5,6,7],['M1','M2','M3','M4','M5','M6','M7'])
plt.show()
```

```python
import matplotlib.pyplot as plt
x=['Monday','Tuesday','Wednesday','Thrusday','Friday','Saturday','Sunday']
y=[10,22,34,23,12,45,60]
plt.bar(x,y,color="Yellow",label='Sales Data')
plt.xlabel("Days of week")
plt.ylabel("Sales per day")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=['Monday','Tuesday','Wednesday','Thrusday','Friday','Saturday','Sunday']
y=[10,22,34,23,12,45,60]
plt.barh(x,y,color="Yellow",label='Sales Data')
plt.xlabel("Sales per day")
plt.ylabel("Days of week")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=['North','South','East','West']
y=[15,40,25,20]
plt.pie(y,labels=x,autopct='%1.1f%%',colors=['orange','lavender','yellow','coral'])
plt.show()
```
